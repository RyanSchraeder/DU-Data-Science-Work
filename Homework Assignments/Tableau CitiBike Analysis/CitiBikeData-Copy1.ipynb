```python
import pandas as pd
import datetime
import numpy as np
import os
```

```python
April_Data = os.path.join("Resources", "JC-201904-citibike-tripdata.csv")
May_Data = os.path.join("Resources", "JC-201905-citibike-tripdata.csv")
June_Data = os.path.join("Resources", "JC-201906-citibike-tripdata.csv")
```

```python
april_data_csv = pd.read_csv(April_Data)
may_data_csv = pd.read_csv(May_Data)
june_data_csv = pd.read_csv(June_Data)

full_Q2_table = pd.concat([april_data_csv, may_data_csv, june_data_csv])
```
